```python
# Importing Data to dataframe from json
import pandas as pd
df1 = pd.read_json ("heal-weheal-userWiseSmallProfile-export.json")

df1 = df1.T.set_index('t', append=True).T
df1
healers = df1.xs('HEALER', axis=1, level=1)
healers = healers.transpose()
healers.head(5)
```

```python
import json
pd.options.display.max_colwidth = 1000
df2 = pd.read_json ("healingSessionsTruncated.JSON")

dataset = pd.DataFrame(df2)
dataset = dataset.transpose()
dataset = dataset[dataset['msgs'].notna()]
dataset['sessionId'] = dataset.index
res = dataset.groupby('sessionId')['msgs'].apply(list).to_dict()
df = pd.DataFrame({'SessionID': list(res.keys()), 'values': list(res.values())}).explode('values').reset_index(drop=True)

lst = []
for index, row in df.iterrows():
    for key, value in row['values'].items():
        lst.append([row['SessionID'], key, value])
new = pd.DataFrame(lst,columns=['SessionID', 'EachTextByUserID', 'value'])


sample = new.iloc[:,:2]
new = new.iloc[:,1:]

lst1 = []
for index, row in new.iterrows():
    for key, value in row['value'].items():
        lst1.append([row['EachTextByUserID'], key, value])
new1 = pd.DataFrame(lst1,columns=['EachTextByUserID','DeepColumns', 'value'])

final = pd.merge(sample,new1, on = 'EachTextByUserID')
final['EachTextByUserID']= final['EachTextByUserID'].str[1:]
#final['store value'] = final['value']
final.head(10)

## FINAL IS THE DATAFRAME AFTER EXTRACTION OF EACH LEVEL OF DICTIONARY IN JSONO FILE WHERE THE MSGS VAUE IS AVAILABLE
```
